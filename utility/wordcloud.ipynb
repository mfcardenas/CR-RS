```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
```

```python
data_path = "/Users/Jack/Documents/Projects/Whatscooking-heroku-/input/df_parsed.csv"
data = pd.read_csv(data_path)
data.head()
```

```python
text = " ".join(str(ingred) for ingred in data.ingredients_parsed)
print ("There are {} words in the combination of all review.".format(len(text)))
```

```python
wordcloud = WordCloud(background_color="white", collocations=True, colormap='viridis', width=1360, height=300).generate(text)
plt.figure( figsize=(20,10), facecolor='k')
plt.imshow(wordcloud)
plt.axis("off")
# plt.tight_layout(pad=0)
# plt.show()
plt.savefig('/Users/Jack/Documents/Projects/Whatscooking-heroku-/input/wordcloud.png', facecolor='k', bbox_inches='tight')
```
